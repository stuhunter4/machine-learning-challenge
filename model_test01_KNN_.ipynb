```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df
```

# Select your features (columns)

```python
# set test features from previous random forest classifier preprocessing
selected_features = df[["koi_fpflag_co", "koi_fpflag_nt", "koi_fpflag_ss", "koi_model_snr",
                        "koi_prad", "koi_duration_err2", "koi_duration_err1", "koi_fpflag_ec",
                        "koi_steff_err1", "koi_prad_err1", "koi_prad_err2", "koi_steff_err2"]]
feature_names = selected_features.columns
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# assign X(data) and y(target)
data = selected_features
target = df["koi_disposition"]
#y = df["koi_disposition"].values.reshape(-1, 1)
print(data.shape, target.shape)
```

```python
# split the data into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
X_train
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train the Model



```python
# loop through different k values to see which has the highest accuracy
# note: we only use odd numbers because we don't want any ties
train_scores = []
test_scores = []
for k in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    train_score = knn.score(X_train_scaled, y_train)
    test_score = knn.score(X_test_scaled, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")

plt.plot(range(1, 20, 2), train_scores, marker='o')
plt.plot(range(1, 20, 2), test_scores, marker='x')
plt.xlabel("k neighbors")
plt.ylabel("Testing Accuracy Score")
plt.show()
```

```python
# given that k: 9 provides the best accuracy where the classifier starts to stabilize
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train_scaled, y_train)
print('k=9 Test Acc: %.3f' % knn.score(X_test_scaled, y_test))
```

```python
# make predictions
predictions = knn.predict(X_test_scaled)
```

```python
# calculate classification report
# precision = true positive / (true positive + false positive)
# recall = true positive / (true positive + false negative)
# f1 = 2*((precision * recall) / (precision + recall)): best at 1, worst at 0
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```
