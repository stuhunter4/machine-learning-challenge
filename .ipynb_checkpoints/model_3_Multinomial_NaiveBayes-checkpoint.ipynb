```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df.drop("koi_disposition", axis=1)
feature_names = selected_features.columns
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# assign X(data) and y(target)
data = selected_features
target = df["koi_disposition"]
#y = df["koi_disposition"].values.reshape(-1, 1)
print(data.shape, target.shape)
```

```python
# split the data into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
X_train
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# create a random forest classifier to automatically calculate feature importance
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train_scaled, y_train)
rf.score(X_test_scaled, y_test)
```

```python
# sort the features by their importance
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
print(X_train_scaled.shape, X_test_scaled.shape)
```

```python
# use tree-based estimator coupled with SelectFromModel meta-transformer...
# ...to compute impurity-based feature importances and discard irrelevant features 
from sklearn.feature_selection import SelectFromModel
sfmodel = SelectFromModel(rf, prefit=True)
X_train_scaled = sfmodel.transform(X_train_scaled)
X_test_scaled = sfmodel.transform(X_test_scaled)
print(X_train_scaled.shape, X_test_scaled.shape)
```

# Train the Model



```python
# create lists to plot the features; Naive Bayes model chosen depends on distribution of data
import numpy as np
x_list = np.arange(0, len(X_test), 1).tolist()
features_list = list(X_test.columns)
print(features_list)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
fig, axs = plt.subplots(len(features_list), figsize=(6, 120))
for x, feature in enumerate(features_list):
    f_list = X_test[f"{feature}"].tolist()
    axs[x].plot(x_list, f_list)
    axs[x].set_title(f"{x}:{feature}")
```

### Based on the given dataset, number of features, and distribution, three Naive Bayes Classifiers will be tested

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model
```

```python
# fit(train) our model using the training data
model.fit(X_train_scaled, y_train)
```

```python
# validate the model using the test data
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model/estimator along with a parameter object containing the values to adjust
from sklearn.model_selection import GridSearchCV
param_grid = {'alpha': [0, 1],
              'fit_prior': [True, False]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
# Train the model with GridSearch estimator...
# this will take the model and try each combination of parameters
grid.fit(X_train_scaled, y_train)
```

```python
# list the best parameters for this dataset, then list the best score
print(grid.best_params_)
print(grid.best_score_)
```

```python
# make predictions with the hypertuned model
predictions = grid.predict(X_test_scaled)
```

```python
# calculate classification report
# precision = true positive / (true positive + false positive)
# recall = true positive / (true positive + false negative)
# f1 = 2*((precision * recall) / (precision + recall)): best at 1, worst at 0
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

# Save the Model

```python
import joblib
filename = 'models/exoplanets_M_NaiveBayes.sav'
joblib.dump(grid, filename)
```
