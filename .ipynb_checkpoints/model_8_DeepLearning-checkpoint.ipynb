```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df.drop("koi_disposition", axis=1)
feature_names = selected_features.columns
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# assign X(data) and y(target)
data = selected_features
target = df["koi_disposition"]
#y = df["koi_disposition"].values.reshape(-1, 1)
print(data.shape, target.shape)
```

```python
# split the data into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
X_train
```

# Pre-processing

Scale the data using the MinMaxScaler, perform some feature selection, label-encode the data set, and one-hot encode

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# create a random forest classifier to automatically calculate feature importance
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train_scaled, y_train)
rf.score(X_test_scaled, y_test)
```

```python
# sort the features by their importance
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
print(X_train_scaled.shape, X_test_scaled.shape)
```

```python
# use tree-based estimator coupled with SelectFromModel meta-transformer...
# ...to compute impurity-based feature importances and discard irrelevant features 
from sklearn.feature_selection import SelectFromModel
sfmodel = SelectFromModel(rf, prefit=True)
X_train_scaled = sfmodel.transform(X_train_scaled)
X_test_scaled = sfmodel.transform(X_test_scaled)
print(X_train_scaled.shape, X_test_scaled.shape)
```

```python
# label-encode data set, converting classes to integer labels
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
y_train_encoded = label_encoder.transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

```python
# one-hot encode the target labels, setting each row's class vlaue to an array
from tensorflow.keras.utils import to_categorical
y_train_categorical = to_categorical(y_train_encoded)
y_test_categorical = to_categorical(y_test_encoded)
y_train_categorical
```

# Create a Deep Learning Model



```python
# create a sequential model
from tensorflow.keras.models import Sequential
model = Sequential()
```

```python
# add first hidden layer, number of inputs and number of nodes specified
from tensorflow.keras.layers import Dense
model.add(Dense(units=100, activation='relu', input_dim=12))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=3, activation='softmax'))
```

```python
# model summary
model.summary()
```

```python
# compile model using a loss function (categorical crossentroy for categorical data), an optimizer, and additional metric
# output layer in this instance is using software for logistic regression (i.e. categorical)
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
             metrics=['accuracy'])
```

```python
# fit(train) the model using the training data
# in this instance, we chose 100 iterations/epochs, shuffle data, and increase detail printed each training cycle
model.fit(X_train_scaled, y_train_categorical,
         epochs=1000, shuffle=True, verbose=2)
```

```python
# test ability to predict new data points (validity) by using the testing data to evaluate the model
model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test_categorical, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# use trained model to make predictions
encoded_predictions = model.predict_classes(X_test_scaled[:10])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted Classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:10])}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# use scikit-learn to grid search the batch size and epochs
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

def create_model():
    fmodel = Sequential()
    fmodel.add(Dense(units=100, activation='relu', input_dim=12))
    fmodel.add(Dense(units=100, activation='relu'))
    fmodel.add(Dense(units=3, activation='softmax'))
    fmodel.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return fmodel
model2 = KerasClassifier(build_fn=create_model, verbose=0)
param_grid = {'batch_size': [20, 60, 100],
              'epochs': [50, 100],
             }
grid = GridSearchCV(estimator=model2, param_grid=param_grid, n_jobs=-1, cv=3)
```

```python
# Train the model with GridSearch estimator...
# this will take the model and try each combination of parameters
grid.fit(X_train_scaled, y_train_categorical)
```

```python
# list the best parameters for this dataset, then list the best score
print(grid.best_params_)
print(grid.best_score_)
```

```python
# make predictions with the hypertuned model
predictions = grid.predict(X_test_scaled)
pred_labels = label_encoder.inverse_transform(predictions)
```

```python
y_pred = pred_labels.tolist()
y_test = list(y_test)
```

```python
pred_df = pd.DataFrame({"Predictions": y_pred, "Actual": y_test})
pred_df.head(10)
```
